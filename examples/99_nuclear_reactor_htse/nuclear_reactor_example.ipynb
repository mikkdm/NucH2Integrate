```python
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# import kaleido


def plot_graph(
    x_data,
    y_data,
    line_name,
    title,
    x_title,
    y_title,
    filename,
    line_color,
    y,
    line_styles,
    fig_show=False,
):
    # Define the figure
    fig = make_subplots(specs=[[{"secondary_y": True}]])  # fig = go.Figure()
    # Add data to figure
    for i in range(0, len(x_data), 1):
        fig.add_trace(
            go.Scatter(
                x=x_data[i],
                y=y_data[i],
                name=line_name[i],
                yaxis=y[i],
                line={"color": line_color[i], "dash": line_styles[i]},
            )
        )

    # Create Layout for the Html

    fig.update_layout(
        title=go.layout.Title(text=title, xref="paper", x=0.5),
        xaxis=go.layout.XAxis(
            title=go.layout.xaxis.Title(
                text=x_title,
            ),
            rangeslider={"visible": True},
        ),
    )

    fig.update_layout(
        xaxis={"domain": [0.05, 0.95]},
        yaxis={"title": {"text": "Power (kW)"}},
        yaxis2={"title": {"text": "Hydrogen Produced (kg/h)"}, "overlaying": "y", "side": "right"},
        yaxis3={
            "title": {"text": "yaxis3 title"},
            "anchor": "free",
            "overlaying": "y",
            "autoshift": True,
        },
        yaxis4={
            "title": {"text": "yaxis4 title"},
            "anchor": "free",
            "overlaying": "y",
            "autoshift": True,
        },
    )

    fig.write_html((filename + ".html"), include_mathjax="cdn")
    # fig.write_image(filename + '.svg')
    # fig.write_image(filename + '.pdf')
    if fig_show:
        fig.show()
```

```python
# os.chdir(r"C:\Users\MIKKDM\Documents\H2Integrate\NucH2INtegrate\examples\99_nuclear_reactor_htse")

x_title = "Time"
y_title = ["Power (MW)", "Hydrogen Produced (kg/h)"]
line_name = ["Electricity Out", "Hydrogen Produced"]
line_color = ["deepskyblue", "red"]
line_styles = ["solid", "dash"]
x_data = [np.arange(8760)] * 2
y_data = [nuclear_elec_out, electrolyzer_h2_out]
title = "Nuclear Reactor Optimization Output"
filename = "nuclear_example_output"
y = ["y1", "y2"]
plot_graph(
    x_data,
    y_data,
    line_name,
    title,
    x_title,
    y_title,
    filename,
    line_color,
    y,
    line_styles,
    fig_show=True,
)
```
