```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import kaleido
import os 
def plot_graph(x_data, y_data, line_name, title, x_title, y_title, filename, line_color, y, fig_show = False):
    # Define the figure
    fig = make_subplots(specs=[[{"secondary_y": True}]]) # fig = go.Figure()
    # Add data to figure
    for i in range(0, len(x_data), 1):
        fig.add_trace(go.Scatter(x=x_data[i], y=y_data[i], name= line_name[i], line_color = line_color[i], yaxis = y[i]))
            
    # Create Layout for the Html
    
    fig.update_layout(
        title=go.layout.Title(
            text= title,
            xref="paper",
            x=0.5
        ),
        xaxis=go.layout.XAxis(
            title=go.layout.xaxis.Title(
                text=x_title,
            ),
            rangeslider=dict(
                visible=True
            )
        )
    )

    fig.update_layout(
    xaxis=dict(
        domain=[0.05, 0.95]
    ),
    yaxis=dict(
        title=dict(
            text="Power (kW)"
        )
    ),
    yaxis2=dict(
        title=dict(
            text="Hydrogen Produced (kg/h)"
        ),
        overlaying="y",
        side="right"
    ),
    yaxis3=dict(
        title=dict(
            text="yaxis3 title"
        ),
        anchor="free",
        overlaying="y",
        autoshift=True
    ),
    yaxis4=dict(
        title=dict(
            text="yaxis4 title"
        ),
        anchor="free",
        overlaying="y",
        autoshift=True
    ))
    
    fig.write_html((filename + '.html'), include_mathjax='cdn')
    fig.write_image(filename + '.svg')
    fig.write_image(filename + '.pdf')
    if fig_show:
        fig.show()
```

```python
os.chdir(r"C:\Users\MIKKDM\Documents\H2Integrate\inl_h2integrate\examples\29_nuclear_reactor")
x_title = "Time"
y_title = ["Power (MW)", "Hydrogen Produced (kg/h)"]
line_name = ["Electricity Out", "Hydrogen Produced"]
line_color = ["deepskyblue", "red"]
x_data = [np.arange(8760)] * 2
y_data = [nuclear_elec_out, electrolyzer_h2_out]
title = "Nuclear Reactor Optimization Output"
filename = "nuclear_example_output"
y = ["y1", "y2"]
plot_graph(x_data, y_data, line_name, title, x_title, y_title, filename, line_color, y)

```
